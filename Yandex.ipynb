<b>2.1 Сколько в датасете суммарно ресторанов и пабов?</b> Ответ: 11345

```python
bill = len(df[df['rubric'] == 'Ресторан']) + len(df[df['rubric'] == 'Бар, паб'])
bill
```

<b>2.2 Какой у них средний чек?</b> Ответ: 2065,0581645256243

```python
restorant = df[df['rubric'] == 'Ресторан']
restorant_bill = restorant['average_bill'].sum() # сумма всех чеков в ресторанах
pub = df[df['rubric'] == 'Бар, паб']
pub_bill = pub['average_bill'].sum() # cумма всех чеков в пабах
(restorant_bill + pub_bill) / bill # средний чек
```

<b> 3. Какая доля заведений “Кофе с собой” действительно имеет фичу кофе на вынос? </b> Ответ: 0.3806228373702422

```python
coffe_away = df[df['rubric'] =='Кофе с собой']
feature = len(coffe_away[coffe_away['coffee_to_go'] == 1])
feature / len(coffe_away)
```

<b>4. Найдите разницу между средним чеком всех кафе в Москве и Санкт-Петербурге. </b> Ответ: 132.53222432978873

```python
moscow = df[df['city'] == 'msk']
moscow_caffe = moscow[moscow['rubric'] == 'Кафе']
bill_moscow_caffe = moscow_caffe['average_bill'].mean()
spb = df[df['city'] == 'spb']
spb_caffe = spb[spb['rubric'] == 'Кафе']
bill_spb_caffe = spb_caffe['average_bill'].mean()
bill_moscow_caffe - bill_spb_caffe
```
